```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Setting color
```

```python
x = np.arange(0,25)
y = x **2
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='purple',lw=1)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='purple',lw=10,alpha=0.4)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='gold',lw=4,ls='-.')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='green',lw=4,ls='steps')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='red',lw=4,ls='-',marker='o')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='c',lw=1,ls='-',marker='o',markersize=10)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='c',lw=1,ls='-',marker='o',markersize=10,
          markerfacecolor='green',markeredgewidth=3,markeredgecolor='yellow')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='purple',lw=1)
axes.set_xlim(0,20)
axes.set_ylim(0,10)
```

```python
fig, ax =plt.subplots(1,2,figsize=(10,4))

ax[0].plot(x, x+3, x, np.exp(x))
ax[0].set_title('Normal scale')

ax[1].plot(x, x+3, x, np.exp(x))
ax[1].set_yscale('log')
ax[1].set_title('Logarithmic scale(y)')

plt.show()
```

```python
fig, a = plt.subplots(figsize=(10,4))

a.plot(x,x+3,x,(x*1.5)+1, lw=3)

a.set_xticks([4,9,14,19,24])
a.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'], fontsize=18)


yticks=[10,20,30,35,40]
a.set_yticks(yticks)
a.set_yticklabels(["$%.1f$" % y for y in yticks], fontsize=18)

plt.show()
```
