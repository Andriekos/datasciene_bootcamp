```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
y = x **2
```

```python
x
```

```python
y
```

# sublpots()

```python
fig, axes = plt.subplots()

axes.plot(x,y,'b')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('1st Plolt')
axes[1].plot(y,x,'r')
axes[1].set_title('2nd Plot')
plt.tight_layout()
```

# figure size, aspect ratio and DPI

```python
fig = plt.figure(figsize=(8,4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(9,3))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
fig, axes = plt.subplots(figsize=(12,6),nrows=1,ncols=2)

axes[0].plot(x, y, 'r')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('title')

axes[1].plot(y,x, 'b')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('title')

plt.tight_layout()
```

```python
fig.savefig('matplotlib.png',dpi=200)
```

#Legend

```python
fig = plt.figure()

a = fig.add_axes([0,0,1,1])

a.plot(x, x**2, label='x**2')
a.plot(x, x+3, label='x+3')
a.legend(loc=0)
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(x,x**2,'b*-',label='x**2')
ax.plot(x,x+3,'g.-',label='x+3')
ax.legend(loc=4)
```

```python
fig, b = plt.subplots(figsize=(8,5))

b.plot(x, x+1,color='blue',alpha=0.5,label='x+1')
b.plot(x, x+2,color='#8B008B',label='x+2')
b.plot(x, x+3,color='#FF8C00',label='x+3')

b.legend(loc=4)
```
