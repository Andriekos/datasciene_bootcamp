```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

# Basic Matplotlib Commands

```python
plt.plot(x,y,'b')
plt.xlabel('X Axis Title')
plt.ylabel('Y Axis Title')
plt.title('Title')
plt.show()
```

# Creating Multiple on Same Canvas

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x, 'g*-')
```

# Matplotlib Object Oriented Method 

```python
#create figure (empty canvas)
fig = plt.figure()

#add set of axes on to figure
axes = fig.add_axes([0.1,0.1,0.8,0.8])
#left, bottom, widht, height (range 0 to 1)
axes.plot(x,y,'b--')
axes.set_xlabel('X labels')
axes.set_ylabel('Y labels')
axes.set_title('Set title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #main axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) #inset axes

#larger figure axes 1
axes1.plot(x,y,'b')
axes1.set_xlabel('X large Axes1')
axes1.set_ylabel('Y large Axes1')
axes1.set_title('Axes1')

#figure axes 2
axes2.plot(y,x,'r')
axes2.set_xlabel('X small Axes2')
axes2.set_ylabel('Y small Axes2')
axes2.set_title('Axes2')
```

```python
fig, axes = plt.subplots()

axes.plot(x,y,'b')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('1st Plolt')
axes[1].plot(y,x,'r')
axes[1].set_title('2nd Plot')
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8,4), dpi=100)
```
