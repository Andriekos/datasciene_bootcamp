```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2 + 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'g')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Chart Title')
plt.show
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'b--')
plt.subplot(1,2,2)
plt.plot(y,x,'y.-')
plt.subplot(1,2,2)
plt.plot(x,y,'r*-')
plt.show()
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y,'r-*')
axes.set_xlabel('X Axes')
axes.set_ylabel('Y Axes')
axes.set_title('Title')
```

```python
fig = plt.figure()

a1 = fig.add_axes([0.1,0.1,0.8,0.8])
a2 = fig.add_axes([0.2,0.4,0.3,0.3])

a1.plot(x,y,'y--')
a1.set_xlabel('X large')
a1.set_ylabel('Y large')
a1.set_title('Large')

a2.plot(y,x,'r')
a2.set_xlabel('X small')
a2.set_ylabel('Y small')
a2.set_title('Small')

plt.show()
```
